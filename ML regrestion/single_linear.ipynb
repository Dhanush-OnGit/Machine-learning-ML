```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression 
```

```python
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:/ML csv/training_data.csv")
```

```python
print(data.columns)
```

```python
x = data['x'].values.reshape(-1,1)
```

```python
y = data['y'].values
```

```python
data_cleaned = data.dropna(subset = ['y'])
```

```python
x = data_cleaned['x'].values.reshape(-1,1)
```

```python
y = data_cleaned['y'].values
```

```python
y_filled = np.nan_to_num(y,nan = np.nanmean(y))
```

```python
y = y_filled
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
y_pred = model.predict(x)
```

```python
mse = mean_squared_error(y,y_pred)
```

```python
r2 = r2_score(y,y_pred)
```

```python
plt.scatter(x,y,label="Data points")
plt.plot(x,y_pred,color="red",label = "Single linear")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Single linear regression")
plt.legend()
plt.show()
```
