```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
```

```python
x = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
scaler = StandardScaler()
```

```python
x_scaler = scaler.fit_transform(x)
```

```python
kmeans = KMeans(n_clusters = 3,random_state = 42)
```

```python
kmeans.fit(x_scaler)
```

```python
cluster = kmeans.predict(x_scaler)
```

```python
x['cluster'] = cluster
```

```python
plt.figure(figsize = (10,6))
```

```python
plt.scatter(x.iloc[:,0],x.iloc[:,1],c= x['cluster'],cmap = 'viridis',s=50)
plt.xlabel('sepal length(standadized)')
plt.ylabel('sepal width(standadized)')
centroids = kmeans.cluster_centers_
centroids_2d = centroids[:,:2]
plt.scatter(centroids_2d[:,0],centroids_2d[:,1],c="red",marker ='x',s=200,label ="centroid")
plt.legend()
plt.show()
```
