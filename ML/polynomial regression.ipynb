```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:/ML csv/Ice_cream_selling_data.csv")
```

```python
print(data.columns)
```

```python
x = data['Temperature (°C)'].values.reshape(-1,1)
```

```python
y = data['Ice Cream Sales (units)'].values
```

```python
poly = PolynomialFeatures(degree = 2)
```

```python
x_poly = poly.fit_transform(x)
```

```python
model = LinearRegression()
```

```python
model.fit(x_poly,y)
```

```python
y_pred = model.predict(x_poly)
```

```python
mse = mean_squared_error(y,y_pred)
```

```python
r2 = r2_score(y,y_pred)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(x,y,label = "data")
x_sort = np.sort(x,axis = 1)
x_sort_poly = poly.transform(x_sort)
y_sort = model.predict(x_sort_poly)
plt.plot(x_sort,y_sort,color = "red",label = "polynomial")
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales (units)')
plt.legend()
plt.show()
```
